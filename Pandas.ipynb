### Pandas DataFrame
- DataFrame is just like a table.Which is going to contain rows and column
- Row can have a Index 
- Column can have a meaning full names

```python
import numpy as np
```

```python
temp=np.random.randint(low=20,high=100,size=[20])
name=np.random.choice(['ravi','kishore','anil','kumar'],20)
av=np.random.choice([10,12,13,34,56],20)
```

```python
a=list(zip(temp,name,av))
```

```python
df=pd.DataFrame(data=a,columns=['temp','name','av'])
```

```python
df
```

```python
type(df)
```

```python
# using dictionaries how we can create dataframe
df=pd.DataFrame({'temp':temp,'name':name,'av':av})
```

##### Indexes

```python
# 
df.set_index('temp')
```

```python
# If we didn't give the "inplace=True" the "df" Dataframe will take temperorly
# change index again give old df dataframe
df.set_index('temp',inplace=True)
```

```python
df
```

```python
#decending index
df.sort_index(axis=0,ascending=False)
```

```python
# I Want to sort the value of av column respectively
df.sort_values(by='av',ascending=False)
```

```python
df.drop(['av'],axis=1) # This is nor perment change.If we want to perment
                       #change we must use "inplace=True"
```

```python
df.head()
```

#### loc and iloc

```python
#index(rows) values returns
df.iloc[[0,1]]
```

```python
# I want to rows no 1 to 2 and column no 1 values
df.iloc[1:3,1]
```

```python
# Play with a boolen type
df.iloc[[True,True,False,True]]
```

```python
df.iloc[0,:]
```

```python
df.head()
```

```python
# #when ever using loc it is basically looking for index value
df.loc[95]
```

```python
df.loc[[39,72,97]]
```

```python
df.loc[[True,True,False]]
```

```python
# In this loc we can also defining some "conditionals"
df.loc[df.av>13]
```

```python
# Some logical computating
df.loc[(df.av>14) | (df.av==10)]
```

#### Merging

```python
d1=pd.DataFrame([['a',1],['b',2]],columns=['col1','number'])
d2=pd.DataFrame([['c',3,'lion'],['d',4,'triger']],columns=['letter','number','animal'])
```

```python
d1
```

```python
d2
```

```python
# I Want to concatenate row wise
pd.concat([d1,d2],axis=0)
```

```python
pd.concat([d1,d2],axis=0,ignore_index=True)
```

```python
pd.concat([d1,d2],axis=1)
```
